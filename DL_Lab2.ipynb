```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout, BatchNormalization, Activation
from tensorflow.keras.utils import to_categorical
```

## 1. To implement a simple feed-forward neural network using keras library. The neural network should have the following specifications:

### Loading the dataset

```python
mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.title(f'Label: {y_train[i]}')
    plt.axis('off')
plt.show()
```

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
y_train = to_categorical(y_train, num_classes=10)
y_test = to_categorical(y_test, num_classes=10)
```

### Creating a feed forward neural network

```python
model = Sequential()
model.add(Flatten(input_shape=(28, 28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.1)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test accuracy: {accuracy * 100:.2f}%')
```

## 3. Modify your feed-forward neural network to include batch normalization layers after each hidden layer.

### Model with Batch Normalization 

```python
model_with_bn = Sequential()
model_with_bn.add(Flatten(input_shape=(28, 28)))
model_with_bn.add(Dense(128, activation='relu'))
model_with_bn.add(BatchNormalization())
model_with_bn.add(Dense(64, activation='relu'))
model_with_bn.add(BatchNormalization())
model_with_bn.add(Dense(10, activation='softmax'))
```

```python
model_with_bn.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history_with_bn = model_with_bn.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.1)
```

```python
loss, accuracy = model_with_bn.evaluate(x_test, y_test)
print(f'Test accuracy: {accuracy * 100:.2f}%')
```

### Model without Batch Normalization

```python
model_without_bn = Sequential()
model_without_bn.add(Flatten(input_shape=(28, 28)))
model_without_bn.add(Dense(128, activation='relu'))
model_without_bn.add(Dense(64, activation='relu'))
model_without_bn.add(Dense(10, activation='softmax'))
```

```python
model_without_bn.compile(loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history_without_bn = model_without_bn.fit(x_train, y_train, epochs=10, batch_size=32, validation_split=0.1)
```

```python
loss, accuracy = model_without_bn.evaluate(x_test, y_test)
print(f'Test accuracy: {accuracy * 100:.2f}%')
```

### Comparison Betwwen both 

```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.plot(history_with_bn.history['accuracy'], label='With BatchNorm')
plt.plot(history_without_bn.history['accuracy'], label='Without BatchNorm')
plt.title('Training Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history_with_bn.history['val_loss'], label='With BatchNorm')
plt.plot(history_without_bn.history['val_loss'], label='Without BatchNorm')
plt.title('Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
plt.show()
```
